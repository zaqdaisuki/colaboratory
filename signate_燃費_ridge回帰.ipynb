```python
import numpy as np 
import pandas as pd

train = pd.read_csv("/content/train.csv")
test = pd.read_csv("/content/test.csv")

display(train)
display(test)
```

```python
def kesson_table(df):
  null_val = df.isnull().sum()
  percent = df.isnull().sum()/len(df)
  kesson_table = pd.concat([null_val, percent], axis=1)
  kesson_table_renamed = kesson_table.rename(columns = {0:"欠損数", 1:"%"})
  return kesson_table_renamed

display(kesson_table(train))
display(kesson_table(test))
```

```python
train["WhatIsData"] = "train"
test["WhatIsData"] = "test"
test["mpg"] = 99999
alldata = pd.concat([train, test], axis=0)
display(alldata)
```

```python
def g_d(df):
  df = pd.get_dummies(df, columns=["origin", "cylinders"], drop_first = True)
  return df

alldata_ = g_d(alldata)
display(alldata_)
```

```python
all_mean = alldata.query('horsepower != "?" and WhatIsData == "train"', engine="python").astype({"horsepower":int}).mean()
# print(all_mean["horsepower"])
alldata_ = alldata.replace({"horsepower":{"?":all_mean["horsepower"]}}).astype({"horsepower":int})
alldata_.head(11)
```
